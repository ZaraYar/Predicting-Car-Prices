```python
import pandas as pd
import numpy as np
 
columns = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', "fuel-system",'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', names=columns)
```

```python

# Select only the columns with continuous values
columns_continuous = ["normalized-losses", "wheel-base", 'length', 'width', 'height', 'curb-weight', "engine-size", 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars_mydataset = cars[columns_continuous]
cars_mydataset
```

# Data Cleaning

```python
cars_mydataset = cars_mydataset.replace('?', np.nan)
cars_mydataset
```

```python
cars_mydataset = cars_mydataset.astype('float')
```

```python
cars_mydataset.isnull().sum()
```

```python
# Because `price` is the column we want to predict, let's remove any rows with missing `price` values.
cars_mydataset = cars_mydataset.dropna(subset=['price'])
```

```python
cars_mydataset.isnull().sum()
```

```python
# Replace missing values in other columns using column means.
cars_mydataset = cars_mydataset.fillna(cars_mydataset.mean())
```

```python
cars_mydataset.isnull().sum()
```

```python
# Normalize all columnns to range from 0 to 1 except the target column.
price = cars_mydataset['price']
cars_mydataset = (cars_mydataset - cars_mydataset.min())/(cars_mydataset.max() - cars_mydataset.min())
cars_mydataset['price'] = price
```

```python
cars_mydataset
```

# Univariate Model

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(features, target, df):
    
    # Randomize order of rows in data frame.
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(rand_df) / 2)
    
    # Select the first half and set as training set.
    # Select the second half and set as test set.
    train_df = rand_df.iloc[0:last_train_row]
    test_df = rand_df.iloc[last_train_row:] 
    
    #train    
    model = KNeighborsRegressor()
    model.fit(train_df[[features]], train_df[target])
    
    # Predict
    test_df["predicted_price"] = model.predict(test_df[[features]])
    mse = mean_squared_error(test_df[target], test_df["predicted_price"])
    rmse = mse**(1/2)
    return rmse

rmse = {}
train_cols = cars_mydataset.columns.drop('price')

# For each column (minus `price`), train a model, return RMSE value
# and add to the dictionary `rmse`.
for col in train_cols:
    rmse_val = knn_train_test(col, 'price', cars_mydataset)
    rmse[col] = rmse_val

# Create a Series object from the dictionary so 
# we can easily view the results, sort, etc
rmse_results_series = pd.Series(rmse)
rmse_results_series.sort_values()

    
    
```

```python

def knn_train_test(train_col, target_col, df):
    np.random.seed(1)
        
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(rand_df) / 2)
    
    # Select the first half and set as training set.
    # Select the second half and set as test set.
    train_df = rand_df.iloc[0:last_train_row]
    test_df = rand_df.iloc[last_train_row:]
    
    k_values = [1,3,5,7,9]
    k_rmses = {}
    
    for k in k_values:
        # Fit model using k nearest neighbors.
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[[train_col]], train_df[target_col])

        # Make predictions using model.
        predicted_labels = knn.predict(test_df[[train_col]])

        # Calculate and return RMSE.
        mse = mean_squared_error(test_df[target_col], predicted_labels)
        rmse = np.sqrt(mse)
        
        k_rmses[k] = rmse
    return k_rmses

k_rmse_results = {}

# For each column (minus `price`), train a model, return RMSE value
# and add to the dictionary `rmse_results`.
train_cols = cars_mydataset.columns.drop('price')
for col in train_cols:
    rmse_val = knn_train_test(col, 'price', cars_mydataset)
    k_rmse_results[col] = rmse_val

k_rmse_results
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for k,v in k_rmse_results.items():
    x = list(v.keys())
    y = list(v.values())
    
    plt.plot(x,y)
    plt.xlabel('k value')
    plt.ylabel('RMSE')
```

# Multivariate Model

```python
def knn_train_test(train_cols, target_col, df):
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(rand_df) / 2)
    
    # Select the first half and set as training set.
    # Select the second half and set as test set.
    train_df = rand_df.iloc[0:last_train_row]
    test_df = rand_df.iloc[last_train_row:]
    
    k_values = [5]
    k_rmses = {}
    
    for k in k_values:
        # Fit model using k nearest neighbors.
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[train_cols], train_df[target_col])

        # Make predictions using model.
        predicted_labels = knn.predict(test_df[train_cols])

        # Calculate and return RMSE.
        mse = mean_squared_error(test_df[target_col], predicted_labels)
        rmse = np.sqrt(mse)
        
        k_rmses[k] = rmse
    return k_rmses

rmse = {}

two_best_features = ['engine-size', 'horsepower']
rmse_val = knn_train_test(two_best_features, 'price', cars_mydataset)
rmse ["two best features"] = rmse_val

three_best_features = ['engine-size', 'horsepower','curb-weight']
rmse_val = knn_train_test(three_best_features, 'price', cars_mydataset)
rmse ["three best features"] = rmse_val

four_best_features = ['engine-size', 'horsepower','curb-weight' ,'highway-mpg']
rmse_val = knn_train_test(four_best_features, 'price', cars_mydataset)
rmse ["four best features"] = rmse_val

five_best_features = ['engine-size', 'horsepower','curb-weight' ,'highway-mpg','width'  ]
rmse_val = knn_train_test(five_best_features, 'price', cars_mydataset)
rmse ["five best features"] = rmse_val


rmse
```
